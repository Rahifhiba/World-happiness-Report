# What are the factors that contribut in making a country the happiest and the saddest in the world ?

```python
def heatmap(df, year):
    """ """
    valid_col = []
    for c in df.columns:
        if not pd.api.types.is_string_dtype(df[c]):
            valid_col.append(c)

    plt.figure(figsize=(10, 8))
    sns.heatmap(df[valid_col].corr(), annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
    plt.title(f'Correlation matrix for data {year}\n')
    plt.show()


```

```python
heatmap(df_15, "2015")
heatmap(df_16, "2016")
heatmap(df_17, "2017")
heatmap(df_18, "2018")
heatmap(df_19, "2019")
```
