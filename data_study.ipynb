## Let's standarize Datas:

```python
# rename columns
df_17 = df_17.rename(columns={
        "Happiness.Rank": "Happiness Rank",
        "Happiness.Score": "Happiness Score",
        "Economy..GDP.per.Capita.": "Economy (GDP per Capita)",
        "Health..Life.Expectancy.": "Health (Life Expectancy)",
        "Trust..Government.Corruption.": "Trust (Government Corruption)",
        "Dystopia.Residual": "Dystopia Residual",
        "Whisker.high": "Lower Confidence Interval",
        "Whisker.low": "Upper Confidence Interval",
})


df_18 = df_18.rename(columns={
    "Country or region": "Country",
    "Overall rank": "Happiness Rank",
    "Score": "Happiness Score",
    "GDP per capita": "Economy (GDP per Capita)",
    "Healthy life expectancy": "Health (Life Expectancy)",
    "Freedom to make life choices": "Freedom",
    "Perceptions of corruption": "Trust (Government Corruption)",


})
df_19 = df_19.rename(columns={
    "Country or region": "Country",
    'Overall rank': 'Happiness Rank',
    "Score": "Happiness Score",
    "GDP per capita": "Economy (GDP per Capita)",
    "Healthy life expectancy": "Health (Life Expectancy)",
    "Freedom to make life choices": "Freedom",
    "Perceptions of corruption": "Trust (Government Corruption)",


})
df_18['Trust (Government Corruption)'].fillna(df_18['Trust (Government Corruption)'].mean(), inplace=True)
```

```python
# df_15 = df_15.rename(columns={
#     'country': 'Country or region',

# })

df_15['Year'] = 2015
df_16['Year'] = 2016
df_17['Year'] = 2017
df_18['Year'] = 2018
df_19['Year'] = 2019

#  z = 1.96 as 95%
df_15['Lower Confidence Interval'] = df_15['Happiness Score'] - 1.96 * df_15['Standard Error']
df_15['Upper Confidence Interval'] = df_15['Happiness Score'] - 1.96 * df_15['Standard Error']
df_15.drop(['Region'], axis=1, inplace=True)
fraction = 0.05
df_18['Standard Error'] = fraction * df_18['Happiness Score']
df_19['Standard Error'] = fraction * df_19['Happiness Score']

df_18['Lower Confidence Interval'] = df_18['Happiness Score'] - 1.96 * df_18['Standard Error']
df_18['Upper Confidence Interval'] = df_18['Happiness Score'] - 1.96 * df_18['Standard Error']

df_19['Lower Confidence Interval'] = df_19['Happiness Score'] - 1.96 * df_19['Standard Error']
df_19['Upper Confidence Interval'] = df_19['Happiness Score'] - 1.96 * df_19['Standard Error']

df_18['Dystopia Residual'] = 1.85 + df_18['Economy (GDP per Capita)'] + df_18['Social support'] + df_18['Freedom'] + df_18['Generosity'] + df_18['Trust (Government Corruption)'] + df_18['Health (Life Expectancy)']
df_19['Dystopia Residual'] = 1.95 + df_19['Economy (GDP per Capita)'] + df_19['Social support'] + df_19['Freedom'] + df_19['Generosity'] + df_19['Trust (Government Corruption)'] + df_19['Health (Life Expectancy)']
```

## Check correlation before combining dataframes

```python
def heatmap(df, year):
    """create heatmap for correlation matrix"""
    valid_col = []
    for c in df.columns:
        if not pd.api.types.is_string_dtype(df[c]) and c != 'Year':
            valid_col.append(c)

    plt.figure(figsize=(10, 8))
    sns.heatmap(df[valid_col].corr(), annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
    plt.title(f'Correlation matrix for data {year}\n')
    plt.show()


```

```python
dfs = [df_15, df_16, df_17, df_18, df_19]
years = [2015, 2016, 2017, 2018, 2019]
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20, 20))
axes = axes.flatten()
for i, (df, y) in enumerate(zip(dfs, years)):
    valid_col = [c for c in df.columns if not pd.api.types.is_string_dtype(df[c]) and c != 'Year']
    sns.heatmap(
        df[valid_col].corr(),
        annot=True,
        cmap='coolwarm',
        linewidths=0.5,
        vmin=-1,
        vmax=1,
        ax=axes[i]
    )
    axes[i].set_title(f'Correlation matrix for data {y}\n')
for j in range(len(dfs), len(axes)):
    axes[j].set_visible(False)
plt.tight_layout()
plt.show()


# heatmap_all_years([df_15, df_16, df_17, df_18, df_19])
```

```python
def get_high_corr(df):
    """get the highest and lowest correlation pairs"""
    valid_col = []
    for c in df.columns:
        if not pd.api.types.is_string_dtype(df[c]):
            valid_col.append(c)
    corr_matrix = df[valid_col].corr()
    corr_pairs = corr_matrix.unstack()

    corr_pairs = corr_pairs[
            (corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)) &
            (corr_pairs.index.get_level_values(0) < corr_pairs.index.get_level_values(1))
        ]
    print('Highest correlation pairs:')
    print(corr_pairs.sort_values(ascending=False).head(4))
    print("*"*30)
    print('Lowest correlation pairs:')
    print(corr_pairs.sort_values(ascending=False).tail(4))

get_high_corr(df_15)
```
